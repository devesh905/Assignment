Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans: In Bokeh, glyphs are the basic visual components used to create plots. They represent shapes, lines, or markers that can be added to a plot to display data. To add glyphs to a Bokeh plot, you can use the figure object's methods like line(), circle(), square(), etc., depending on the type of glyph you want to add.

```python
# example
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 5, 3]

plot = figure(title='Bokeh Plot with Line and Circle Glyphs', x_axis_label='X-axis', y_axis_label='Y-axis')

plot.line(x_values, y_values, line_color='blue', line_width=2, legend_label='Line Glyph')

plot.circle(x_values, y_values, fill_color='red', size=10, legend_label='Circle Glyph')

show(plot)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


In Bokeh, you can customize the appearance of a plot, including the axes, title, and legend, by modifying various properties of the figure object.
Axis Labels: You can set the labels for the x-axis and y-axis using the x_axis_label and y_axis_label properties of the figure object.
Plot Title: You can set the title of the plot using the title property of the figure object.
Legend: If you have multiple glyphs on the plot, you can add a legend to differentiate them. You can customize the legend using properties like legend_label, legend_location, etc.

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

A Bokeh server is a server that allows for the creation of interactive Bokeh plots that can be updated in real-time. Bokeh provides a server framework that enables the development of dynamic, interactive web applications. With Bokeh server, you can create web applications that respond to user interactions, streaming data, or scheduled updates, allowing you to build interactive data dashboards and visualizations.


```python
from bokeh.io import curdoc
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.layouts import column
import numpy as np

source = ColumnDataSource(data={'x': [], 'y': []})

plot = figure(plot_width=400, plot_height=400, title='Dynamic Plot')
plot.circle(x='x', y='y', source=source, size=10, color='blue')

def update():
    new_data = {'x': np.random.random(100), 'y': np.random.random(100)}
    source.data = new_data

curdoc().add_periodic_callback(update, 1000)

curdoc().add_root(column(plot))
```
