Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Ans: A scatter plot  uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point

```python
import numpy as np 
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

Ans: The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots in rows and columns, making it easier to compare different datasets or visualize various aspects of the same data. 

```python
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig = plt.figure()

plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='o', color='b')
plt.title('Line 1')


plt.subplot(2, 2, 2)
plt.plot(x, y2, marker='s', color='g')
plt.title('Line 2')


plt.subplot(2, 2, 3)
plt.plot(x, y3, marker='^', color='r')
plt.title('Line 3')


plt.subplot(2, 2, 4)
plt.plot(x, y4, marker='d', color='purple')
plt.title('Line 4')

# Adjust layout
plt.tight_layout()

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

ans: A bar plot is a graph that presents categorical data with rectangular bars. The lengths of the bars are proportional to the values they represent

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# bar plot 
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profit of Companies')
plt.show()
```

```python
# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Profit of Companies')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a whisker plot, displays a summary of a set of data values. It shows important statistics, such as the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum, using a box and whiskers

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.show()
```
